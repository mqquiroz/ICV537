```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
#Función de Fuerza (externo)
g = 9.81 #(m/s2)
#Parámetros:
m = 68.1 #(kg)
c = 12.5 #(kg/s)

#Función analítica
def vel(t):
    v = g*m/c * (1 - np.exp(-(c/m)*t))
    return v

#definimos un array para el tiempo
t0 = 0 #(s)
tend = 25 #(s)
dt   = 0.01
t = np.arange(t0,tend+dt,dt)

v =vel(t)

#Modelo Numérico en Diferencias Finitas
h = [1, 3, 5, 6]
hc = ['r','b','g','y']

for i in range(len(h)):
  th = np.arange(t0,tend+h[i],h[i]) # tiempos (s)
  velocidad = np.zeros(len(th))
  for j in range(len(th)-1):
    velocidad[j+1] = velocidad[j] + (g -(c/m)*velocidad[j])*(th[j+1]-th[j])
  plt.plot(th,velocidad,hc[i],label='h ='+str(h[i])+' s')



plt.plot(t,v,'k',label='Exacta')
plt.axhline(y=53.44,c='m',linestyle='--')
plt.grid()
plt.xlim(0,1.1*t[-1])
plt.ylim(0,1.25*v[-1])
plt.xlabel('tiempo (s)')
plt.ylabel('velocidad (m/s)')
plt.legend()
plt.show()
```
